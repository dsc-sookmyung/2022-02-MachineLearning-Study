```python
#매트릭스 미사용, 가중치를 하나씩 선언 -> w1, w2, w3
import tensorflow as tf
# data and label
x1=[73., 93., 89., 96., 73.]
x2=[80., 88., 91., 98., 66.]
x3=[75., 93., 90., 100., 70.]
y=[152., 185., 180., 196., 142.]

#random weights
w1=tf.Variable(tf.random.normal([1]))
w2=tf.Variable(tf.random.normal([1]))
w3=tf.Variable(tf.random.normal([1]))
b=tf.Variable(tf.random.normal([1]))

learning_rate=0.000001

for i in range(1001):
    #경사하강 알고리즘을 사용하여 최솟값 W을 구함
    with tf.GradientTape() as tape:
        hypothesis=w1*x1 + w2*x2 + w3*x3 + b
        cost=tf.reduce_mean(tf.square(hypothesis - y)) #오차 제곱의 평균 값으로 비용을 정의
    #calculates the gradients of the cost
    w1_grad, w2_grad, w3_grad, b_grad=tape.gradient(cost,[w1,w2,w3,b]) #알고리즘에 cost함수와 인자(w1 ...)를 넣어 기울기 값을 저장
    
    #update w1,w2,w3,b
    w1.assign_sub(learning_rate*w1_grad)
    w2.assign_sub(learning_rate*w2_grad)
    w3.assign_sub(learning_rate*w3_grad)
    b.assign_sub(learning_rate*b_grad)
    
    if i%50==0:
        print("{:5}|{:12.4f}".format(i,cost.numpy()))
```

```python
#매트릭스 사용, 모든 가중치를 W 하나로 표현
import numpy as np
data=np.array([ 
    #x1,  x2,  x3,  y
    [73., 80., 75., 152. ],
    [93., 88., 93., 185. ],
    [89., 91., 90., 180. ],
    [96., 98., 100., 196. ],
    [73., 66., 70., 142. ]
], dtype=np.float32)

#slice data
x=data[:,:-1] #[전체 행, 마지막 열 제외] -> x1, x2, x3열의 모든 행 -> 5*3 행렬
y=data[:,[-1]] #[전체 행, 마지막 열] -> y열의 모든 행 -> 5*1 행렬

W=tf.Variable(tf.random.normal([3,1])) #[X의 칼럼의 수, 출력의 칼럼의 수]
b=tf.Variable(tf.random.normal([1]))

learning_rate=0.000001

#hypothesis, prediction function
def predict(x):
    return tf.matmul(x,W)+b

n_epochs=2000
for i in range(n_epochs+1):
    #record thr gradient of the cost function
    with tf.GradientTape() as tape:
        cost=tf.reduce_mean((tf.square(predict(x)-y)))
    
    #calculates the gradients of the loss
    W_grad, b_grad=tape.gradient(cost,[W,b])
    
    #updates parameters (W and b)
    W.assign_sub(learning_rate*W_grad)
    b.assign_sub(learning_rate*b_grad)
    
    if i%100==0:
        print("{:5}|{:10.4f}".format(i,cost.numpy()))
```
