## Lec 3. How to minimize cost?

üìå **Goal : DataÎ•º ÌÜµÌï¥ CostÍ∞Ä ÏµúÏÜåÍ∞íÏù¥ ÎêòÎäî W(+ b) Ï∞æÍ∏∞** 

- Simplified Hypothesis >>  H(x) = Wx (b ÏÉùÎûµ) ‚Üí CostÎèÑ b ÏÉùÎûµ



- data setÏùÑ Ïù¥Ïö©Ìï¥ costÍ∞íÏù¥ ÏµúÏ†ÄÏ†êÏù¥ ÎêòÎäî W, bÍ∞íÏùÑ Ï∞æÏùå
- Ïª¥Ìì®ÌÑ∞Îäî Ïñ¥ÎñªÍ≤å ÏµúÏ†ÄÏ†êÏùÑ Ï∞æÎÇò? ‚Üí Gradient descent algorithm Ïù¥Ïö©
    - Î≥ÄÏàòÍ∞Ä 2Í∞ú Ïù¥ÏÉÅÏùº ÎïåÎèÑ Ï∞æÍ∏∞ Í∞ÄÎä•

üìå**Gradient descentÍ∞Ä ÎèôÏûëÌïòÎäî Î∞©Ïãù**

1. W, bÏùò initial Í∞íÏùÑ Ï†ïÌï® : Î≥¥ÌÜµ 0, 0ÏóêÏÑú ÏãúÏûë

2. cost(W, b)Í∞íÏùÑ costÍ∞Ä ÏµúÏÜåÌôîÎêòÎäî Î∞©Ìñ•ÏúºÎ°ú W, bÍ∞í ÏßÄÏÜçÏ†ÅÏúºÎ°ú Ï°∞Ï†ïÌï¥ÎÇòÍ∞ê
    - ÌäπÏ†ï Í∞í(W)ÏóêÏÑú gredient Íµ¨Ìï® 
    - W - (W * Gradient)Î•º Íµ¨Ìï¥ Îã§Ïãú WÏóê Ï†ÄÏû• ‚ÜíÏù¥Îü¨Î©¥ costÍ∞Ä ÏµúÏÜåÍ∞Ä ÎêòÎäî ÏßÄÏ†êÏóê ÎèÑÎã¨. Í∑∏Îïå WÍ∞í Ïïå Ïàò ÏûàÏùå
    
    ‚úÖ **Ïó¨Í∏∞ÏÑú WÎäî Í∏∏Ïù¥Í∞Ä ÏïÑÎãàÎùº XÏ∂ïÏúºÎ°ú Îì§Ïñ¥Í∞ê! Ìó∑Í∞àÎ¶¨ÏßÄ Îßê Í≤É. YÏ∂ïÏùÄ costÏûÑ**

3. minimumÏóê ÎèÑÎã¨ÌñàÎã§Í≥† ÌåêÎã®Îê† Îïå ÍπåÏßÄ Î∞òÎ≥µ

- GradientÎäî cost(W)Î•º ÎØ∏Î∂ÑÌï¥ÏÑú Íµ¨Ìï† Ïàò ÏûàÏùå (Ïù¥Îïå 2mÏúºÎ°ú ÎÇòÎàî - ÎØ∏Î∂Ñ Ïãú ÍπîÎÅîÌïòÍ≤å ÌïòÎ†§Í≥†)





- WÏóê ÎåÄÌï¥ Ìé∏ÎØ∏Î∂Ñ Ìï¥ Gradient Í∞í Íµ¨Ìï®. Îî∞ÎùºÏÑú ÏúÑ ÏàòÏãùÏù¥ gradient descent ÌëúÌòÑ
- Ïù¥Îïå Œ±Í∞í == Learning rate(ÌïôÏäµÎ•†), WÏùò update Ï†ïÎèÑ Í≤∞Ï†ï

üìå**Gradient descent algorithm Ï†ïÎ¶¨_ÏµúÏ¢Ö_ÏµúÏµúÏ¢Ö..**

- W(Weight)Í∞íÏùÑ ÏßÄÏÜçÏ†ÅÏúºÎ°ú ÏóÖÎç∞Ïù¥Ìä∏ Ìï¥ÎÇòÍ∞ÄÎäî ÏïåÍ≥†Î¶¨Ï¶ò
- Í∑∏ Ï†Ñ WÏóêÏÑú cost(w)' * Œ±Î•º ÎπºÏ§Ä Í∞íÏùÑ Îã§Ïãú WÎ°ú Í∞±Ïã†
- Í∑∏Îüº costÏùò ÏàúÍ∞Ñ Î≥ÄÌôîÏú®Ïùò Í∞íÏù¥ 0Ïóê Í∞ÄÍπåÏö∏ Îïå (= costÍ∞Ä ÏµúÏÜåÏùº Îïå) WÏùò Î≥ÄÌôî ÎòêÌïú Í±∞Ïùò ÏóÜÏùÑ Í≤É => W ÌôïÏ†ï

- Ïù¥Îïå, Ï∞®ÏõêÏù¥ ÌôïÏû•Îê† Í≤ΩÏö∞ ÎèÑÎã¨Ìïú cost ÏµúÏ†ÄÏ†êÏù¥ global minimumÏù¥ ÏïÑÎãàÎùº local minimumÏùºÏàòÎèÑ ÏûàÏùå (ÏàòÌïôÏãúÍ∞ÑÏóê Î∞∞Ïö¥ Î∂ÄÎ∂Ñ ÏµúÏÜüÍ∞íÏù∏Í∞ÄÎûë ÎπÑÏä∑ÌïúÎìØ)
    - Convex function : local min == global minÏù∏ Í≤ΩÏö∞ (local minÏù¥ 2Í∞ú Ïù¥ÏÉÅ Ï°¥Ïû¨X)

# Ïã§Ïäµ

```python
import tensorflow as tf
import numpy as np
#tf.enable_eager_execution()
#Ï¶âÏãúÏàòÌñâ
```

‚úÖtf.reduce_mean()
- ÌèâÍ∑† Íµ¨ÌïòÎäî Î©îÏÜåÎìú
- reduce -> Ï∞®Ïõê(Rank)Í∞Ä ÌïòÎÇò Ï§ÑÏñ¥Îì§Î©¥ÏÑú ÌèâÍ∑† Íµ¨Ìï®

‚úÖtf.square()
- ÎÑòÍ≤®Î∞õÏùÄ Í∞í Ï†úÍ≥±Ìï®

‚úÖError == ÏòàÏ∏°Í∞í - Ïã§Ï†úÍ∞í

‚úÖCost == (ÏòàÏ∏°Í∞í - Ïã§Ï†úÍ∞í)^2Ïùò ÌèâÍ∑†

- Cost functionÏùÑ tensorflow ÏΩîÎìúÎ°ú Ï†ïÏùò
    - cost = tf.reduce_mean(tf.square(hypothesis - y_data))

# Gradient descent (Í≤ΩÏÇ¨ÌïòÍ∞ï ÏïåÍ≥†Î¶¨Ï¶ò)
: Cost functionÏóêÏÑú CostÎ•º minimizeÌïòÎäî W, bÎ•º Ï∞æÎäî ÏïåÍ≥†Î¶¨Ï¶ò 


- GradientTape()
    - Ï£ºÎ°ú with Íµ¨Î¨∏Í≥º ÏÑ∏Ìä∏Î°ú ÏÇ¨Ïö©Îê®
    - with Íµ¨Î¨∏ Ïïà Î≥ÄÏàòÎì§Ïùò Î≥ÄÌôî Ï†ïÎ≥¥Î•º tapeÏóê Í∏∞Î°ùÌï®
- gradient(A, [b.c])
    - tapeÏùò Î©îÏÜåÎìú
    - Ìï®Ïàò(A)Ïóê ÎåÄÌï¥ Î≥ÄÏàòÎì§([b,c])Ïóê ÎåÄÌïú Í∞úÎ≥Ñ ÎØ∏Î∂ÑÍ∞íÏùÑ Íµ¨Ìï¥ tupleÎ°ú Î∞òÌôò 
- A.assign_sub(B) 
    - A = A - B


```python
#H(x) = Wx + b
# input, output Í∞íÏù¥ ÏùºÏπòÌïòÎäî dataÍ∞í (ÏïåÎßûÍ≤å ÎÇòÏò§Î†§Î©¥ W = 1, b = 0Ïù¥ ÎÇòÏôÄÏïº Ìï®Ïù¥ ÏòàÏ∏° Í∞ÄÎä•ÌïòÎã§.)
x_data = [1, 2, 3, 4, 5]
y_data = [1, 2, 3, 4, 5]

#tensorflowÏùò variableÎ°ú W, b Ï¥àÍ∏∞Í∞í Ï†ïÏùòÌïòÍ∏∞ (ÏûÑÏùòÏùò Í∞í)
W = tf.Variable(2.9)
b = tf.Variable(0.5)

#Gradient descentÎ•º tensorflowÎ°ú Íµ¨ÌòÑÌïòÍ∏∞ - ÏùºÎ∞òÏ†ÅÏúºÎ°ú GradientTapeÎ•º ÏÇ¨Ïö©Ìï¥ Ï£ºÎ°ú Íµ¨ÌòÑÌï®

#Learning_rate initialize : Í∞Å Î≥ÄÏàòÏóê ÎåÄÌï¥ Íµ¨Ìïú gradient Í∞íÏùÑ ÏñºÎßàÎÇò Î∞òÏòÅÌï†ÏßÄ Í≤∞Ï†ï(ÏûëÏùÑÏàòÎ°ù Ï†êÏßÑÏ†Å Î≥ÄÌôî)
learning_rate = 0.01

for i in range(100): #W, b ÏóÖÎç∞Ïù¥Ìä∏
    #Gradient descent (tapeÏóê Î≥ÄÏàò W, bÏùò Ï†ïÎ≥¥ Í∏∞Î°ù)
    with tf.GradientTape() as tape:
        hypothesis = W * x_data + b
        cost = tf.reduce_mean(tf.square(hypothesis - y_data))
    #gradient - costÎùºÎäî Ìï®ÏàòÏóê ÎåÄÌï¥ W, bÏùò Í∏∞Ïö∏Í∏∞ Í∞ÅÍ∞Å Íµ¨Ìï®
    W_grade, b_grade = tape.gradient(cost, [W,b])
    #assing_sub
    W.assign_sub(learning_rate * W_grade)
    b.assign_sub(learning_rate * b_grade)
    if i % 10 == 0: #10Î≤à Îã®ÏúÑÎ°ú W, b, const Î≥ÄÌôî ÌôïÏù∏ 
        print("{:5}|{:10.4f}|{:10.4f}|{:10.6f}".format(i, W.numpy(), b.numpy(), cost))

#Ïù¥Î†áÍ≤å epoch 100Î≤à Ïù¥ÌõÑ ÏµúÏ¢Ö W, bÍ∞í Í∞±Ïã†Îê® -> hypothesis Í≤∞Ï†ï
```

W
- Ï£ºÏñ¥ÏßÑ Ï¥àÍ∏∞ Í∞í (2.9)Î∂ÄÌÑ∞ ÏãúÏûë ~ Ï†êÏ†ê Í∞êÏÜåÌï¥ 1.0ÏúºÎ°ú ÏàòÎ†¥

b
- 0.5ÏóêÏÑú ÏãúÏûë ~ Ï†êÏ†ê Í∞êÏÜåÌï¥ -0.0 Ï¶â, Í±∞Ïùò 0ÏúºÎ°ú ÏàòÎ†¥

cost
- costÎäî Í∞íÏù¥ ÏûëÏùÑÏàòÎ°ù Ï¢ãÏùå!! (100Î≤à Ïù¥ÏÉÅ ÏàòÌñâ Ïãú 0Ïóê Í∞ÄÍπåÏö¥ Í∞í)
- costÍ∞Ä 0Ïóê Í∞ÄÍπùÎã§ = Î™®Îç∏(Í∞ÄÏÑ§)Ïù¥ Ïã§Ï†ú dataÏôÄ Í±∞Ïùò ÏùºÏπòÌïòÍ≤å Îê® (Ïò§Ï∞® ÏûëÏïÑÏßê) = Ïã§Ï†úÍ∞í ÏòàÏ∏°Ïóê good 

```python
# predict (Ïù¥Î†áÍ≤å ÏôÑÏÑ±Îêú hypothesisÎ•º Í∞ÄÏßÄÍ≥† ÏÉàÎ°úÏö¥ dataÎ°ú ÏòàÏ∏°Ìï¥Î≥¥Í∏∞)
# ÏÉàÎ°úÏö¥ data (x = 5, 2.5)Î•º ÎÑ£Ïñ¥ÎèÑ y_dataÍ∞Ä Í∑∏ÎåÄÎ°ú Ïûò Ï∂úÎ†•Îê®?
print(W * 5 + b)
print(W * 2.5 + b)
```

# cost minmize Íµ¨ÌòÑ
: cost(W) = W(x)Î°ú Îã®ÏàúÌôîÎêú ÏÉÅÌÉú

‚úÖ np.linspace (start, end, interval)
- start ~ end ÏÇ¨Ïù¥Î•º intervalÎßåÌÅº ÎÇòÎàà Í∞í Ï∞®Î°ÄÎ°ú ÎèÑÏ∂ú



```python
# 1. cost(W) Íµ¨ÌòÑ (in python) 

import numpy as np

X = np.array([1, 2, 3])
Y = np.array([1, 2, 3])

#cost(W) Ï†ïÏùò
def cost_func(W, X, Y):
    c = 0
    for i in range(len(X)):
        c += (W * X[i] - Y[i])**2 #c == ÏòàÏ∏°Í∞í - Ïã§Ï†úÍ∞íÏùò Ï†úÍ≥±Ïùò Ìï© 
    #ÏµúÏ¢Ö costÍ∞í Î∞òÌôò
    return c / len(X)

for feed_W in np.linspace(-3, 5, num=15):
    curr_cost = cost_func(feed_W, X, Y)
    print("{:6.3f} | {:10.5f}".format(feed_W, curr_cost))
    
#costÎäî W Í¥ÄÎ†® Ìï®ÏàòÏù¥ÎØÄÎ°ú W Í∞íÏóê Îî∞Îùº Î≥ÄÌôî
```

```python
# 2. cost(W) Íµ¨ÌòÑ (in Tensorflow)

X = np.array([1, 2, 3])
Y = np.array([1, 2, 3])

def cost_func(W, X, Y):
    hypothesis = X * W #ÏòàÏ∏°Í∞í
    return tf.reduce_mean(tf.square(hypothesis - Y))

W_values = np.linspace(-3, 5, num=15)
cost_values = []

for feed_W in W_values:
    curr_cost = cost_func(feed_W, X, Y)
    cost_values.append(curr_cost)
    print("{:6.3f} | {:10.5f}".format(feed_W, curr_cost))
```

‚úÖ tf.Variable()

- ÌååÎùºÎØ∏ÌÑ∞Î•º Î≥ÄÏàòÎ°ú ÎßåÎì§Ïñ¥ÏÑú Ìï†ÎãπÌï¥ Ï†ïÏùòÌï®

‚úÖ tf.random_normal(Î™®Ïñë Î∞è Í∞úÏàò, start, end)

- Ï†ïÍ∑ú Î∂ÑÌè¨Î•º Îî∞Î•¥Îäî (start~end Î≤îÏúÑ ÎÇ¥) random numberÎ•º nÍ∞ú ÏßúÎ¶¨ Î≥ÄÏàòÎ°ú ÎßåÎì¶

```python
# 3. gradient descent Íµ¨ÌòÑ (in Tensorflow)

#ÏΩîÎìú Ïó¨Îü¨Î≤à ÏàòÌñâÌï† Îïå ÎèôÏùºÌïú Ï°∞Í±¥ Î≥¥Ïû• - random seedÎ•º 0ÏúºÎ°ú Ï¥àÍ∏∞Ìôî
tf.random.set_seed(0)

x_data = [1., 2., 3., 4.]
y_data = [1., 3., 5., 7.]

#-100~100 ÏÇ¨Ïù¥ ÎûúÎç§Ìïú Î≥ÄÏàò 1Í∞ú ÎßåÎì§Ïñ¥ÏÑú WÏóê Ìï†ÎãπÌï¥ Ï†ïÏùòÌï®
#W = tf.Variable(tf.random.normal((1,), -100., 100.))
#W = tf.Variable([5.0])
W = tf.Variable([100.0])

for step in range(300):
    hypothesis = W * X
    cost = tf.reduce_mean(tf.square(hypothesis - Y)) #cost func Íµ¨ÌòÑ

    #gradient descent Íµ¨ÌòÑ
    alpha = 0.01 #learning_rate
    gradient = tf.reduce_mean(tf.multiply(tf.multiply(W, X) - Y, X))# Cost(W)' (Í∏∞Ïö∏Í∏∞)
    descent = W - tf.multiply(alpha, gradient)
    #WÏóê Í≥ÑÏÇ∞Ìïú Í∞í Ìï†Îãπ (+=)
    W.assign(descent)
    
    #10Î≤àÎßàÎã§ cost, W Í∞í Ï§ëÍ∞Ñ Ï∂úÎ†•
    if step % 10 == 0:
        print('{:5} | {:10.4f} | {:10.6f}'.format(step, cost.numpy(), W.numpy()[0]))
```

costÍ∞í -> 0ÏúºÎ°ú ÏàòÎ†¥

WÍ∞í -> 1Î°ú ÏàòÎ†¥ ÌôïÏù∏ Í∞ÄÎä•

**WÏóê Ï≤òÏùåÏóê Ïñ¥Îñ§ Í∞íÏùÑ ÎÑ£ÎçîÎùºÎèÑ Í∞ôÏùÄ Í≤∞Í≥ºÍ∞Ä ÎÇòÏò¥**
